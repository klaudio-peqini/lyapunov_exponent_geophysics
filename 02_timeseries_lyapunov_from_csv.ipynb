# Time series Lyapunov exponent from CSV

This notebook shows how to load a column from a CSV and compute the Lyapunov exponent.
You can point it to your own data under `data/raw/`.


```python
import numpy as np
import pandas as pd
from lyapunov_geo.io import CSVLoadConfig, load_series_from_csv
from lyapunov_geo.lyapunov import SeriesLyapunovConfig, lyapunov_exponent_series

# Example: generate and save a synthetic series (replace with your real CSV)
t = np.linspace(0, 200, 5000)
x = np.sin(t) + 0.05*np.random.default_rng(0).normal(size=t.size)
df = pd.DataFrame({'x': x})
df.to_csv('synthetic.csv', index=False)

series = load_series_from_csv(CSVLoadConfig(path='synthetic.csv', column='x'))
le = lyapunov_exponent_series(series, SeriesLyapunovConfig(eps=0.02, l=10, max_pairs=500, min_separation=10, seed=0))
print('Estimated LE:', le)
```
