# Quickstart: bifurcation diagrams

This notebook generates a bifurcation diagram + Lyapunov curves using the package API.


```python
from lyapunov_geo.bifurcation import BifurcationConfig, logistic_bifurcation, plot_bifurcation
from lyapunov_geo.lyapunov import SeriesLyapunovConfig

cfg = BifurcationConfig(
    steps=1500,
    last=300,
    rounding=4,
    le=SeriesLyapunovConfig(eps=0.01, l=5, max_pairs=500, min_separation=10, seed=0),
)

r, steady, le_f, le_s = logistic_bifurcation(2.8, 4.0, 0.01, x0=0.2, cfg=cfg)
plot_bifurcation(r, steady, le_f, le_s, xlabel='r', outpath='Lyapunov_logistic_map_notebook.pdf', title='Logistic map')
print('Wrote Lyapunov_logistic_map_notebook.pdf')
```
